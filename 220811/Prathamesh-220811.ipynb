```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import yfinance as yf
```

```python
stock_data = yf.download("TATASTEEL.NS", period="3mo")
```

```python
sma = stock_data["Close"].rolling(window=3).mean()
ema = stock_data["Close"].ewm(span=3).mean()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(stock_data.index, stock_data["Close"], label="Close Price")
plt.plot(stock_data.index, sma, label="SMA (3)")
plt.plot(stock_data.index, ema, label="EMA (3)")
plt.xlabel("Date")
plt.ylabel("Price")
plt.title("Moving Average Indicator")
plt.legend()
plt.grid(True)
plt.show()
```
